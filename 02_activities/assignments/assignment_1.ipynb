```python

# Import necessary libraries
from dotenv import load_dotenv
from pathlib import Path
import os
from glob import glob

# Define the absolute path to the .env file
env_path = Path('C:/Users/david/my_directory/production/01_materials/labs/.env')

# Load the .env file
load_dotenv(dotenv_path=env_path)

# Retrieve and verify environment variables
PRICE_DATA = os.getenv("PRICE_DATA")
print(f"PRICE_DATA is set to: {PRICE_DATA}")

# Verify other environment variables
LOG_DIR = os.getenv("LOG_DIR")
print(f"LOG_DIR is set to: {LOG_DIR}")

# Use PRICE_DATA to locate Parquet files
price_data_path = Path(PRICE_DATA).resolve()
print(f"Absolute path for PRICE_DATA: {price_data_path}")

# Check if the PRICE_DATA path exists and is a directory
if not price_data_path.is_dir():
    raise NotADirectoryError(f"The path specified by 'PRICE_DATA' does not exist or is not a directory: {price_data_path}")
else:
    print(f"The directory exists: {price_data_path}")

# Use glob to find all Parquet files in the PRICE_DATA directory
parquet_files = list(price_data_path.glob("*/*.parquet"))

# Verify that Parquet files are found
if not parquet_files:
    print("No Parquet files found in the PRICE_DATA directory.")
else:
    print(f"Found {len(parquet_files)} Parquet files:")
    for file in parquet_files[:5]:  # Display first 5 files
        print(f"- {file}")
```

```python
import dask.dataframe as dd
from pathlib import Path
import os
```

```python
# Read all Parquet files recursively within PRICE_DATA
dd_px = dd.read_parquet(str(price_data_path), engine='pyarrow', recursive=True)

# Set 'ticker' as the index if it's a column in your DataFrame
if 'ticker' in dd_px.columns:
    dd_px = dd_px.set_index("ticker")

# Display the first few rows to verify
dd_px.head()
```

```python
len(dd_px)
```

For each ticker and using Dask, do the following:

+ Add lags for variables Close and Adj_Close.
+ Add returns based on Adjusted Close:
    
    - `returns`: (Adj Close / Adj Close_lag) - 1

+ Add the following range: 

    - `hi_lo_range`: this is the day's High minus Low.

+ Assign the result to `dd_feat`.

(4 pt)

```python
import numpy as np 

dd_feat = (dd_px.groupby('ticker', group_keys=False).apply(
    lambda x: x.assign(Close_lag_1 = x['Close'].shift(1))
)
           )
```

```python
dd_feat = (dd_px.groupby('ticker', group_keys=False).apply(
    lambda x: x.assign(Adj_Close_lag_1  = x['Adj Close'].shift(1))
).assign(
    returns = lambda x: x['Adj Close'] / x ['Adj_Close_lag_1'] - 1
).assign(
    hi_lo_range = lambda x: x['High'] - x['Low']
))
```

```python
len(dd_feat)
```

```python
dd_feat.head()
```

+ Convert the Dask data frame to a pandas data frame. 
+ Add a rolling average return calculation with a window of 10 days.
+ *Tip*: Consider using `.rolling(10).mean()`.

(3 pt)

```python
# Write your code below.

dd_feat = (dd_feat.groupby('ticker', group_keys=False).apply(
    lambda x: x.assign(
        rolling_return_10d = lambda x : x['returns'].rolling(window=10).mean()
    )
)
)
```

```python
dd_feat.head()
```

```python
# Converting the dask dataframe to pandas using compute()
pd_feat = dd_feat.compute()
```

```python
pd_feat.head()
```

```python
pd_feat.sample(n=10)
```
